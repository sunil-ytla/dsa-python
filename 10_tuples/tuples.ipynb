```python
import timeit

# Create a list and a tuple with the same elements
lst = list(t)
tpl = t

# Define a function to iterate over a list
def iterate_list():
    for _ in lst:
        pass

# Define a function to iterate over a tuple
def iterate_tuple():
    for _ in tpl:
        pass

# Measure the time it takes to iterate over the list
list_time = timeit.timeit(iterate_list, number=1000000)

# Measure the time it takes to iterate over the tuple
tuple_time = timeit.timeit(iterate_tuple, number=1000000)

print(f"List iteration time: {list_time}")
print(f"Tuple iteration time: {tuple_time}")
```

```python
import timeit

# Define a function to iterate over a list comprehension
def iterate_list_comprehension():
    [x for x in lst]

# Define a function to iterate over a tuple comprehension
def iterate_tuple_comprehension():
    tuple(x for x in tpl)

# Measure the time it takes to iterate over the list comprehension
list_comprehension_time = timeit.timeit(iterate_list_comprehension, number=1000000)

# Measure the time it takes to iterate over the tuple comprehension
tuple_comprehension_time = timeit.timeit(iterate_tuple_comprehension, number=1000000)

print(f"List comprehension iteration time: {list_comprehension_time}")
print(f"Tuple comprehension iteration time: {tuple_comprehension_time}")
```

Elementwise Sum

Create a function that takes two tuples and returns a tuple containing the element-wise sum of the input tuples.

Example

    tuple1 = (1, 2, 3)
    tuple2 = (4, 5, 6)
    output_tuple = tuple_elementwise_sum(tuple1, tuple2)
    print(output_tuple)  # Expected output: (5, 7, 9)




```python
def tuple_elementwise_sum(tuple1, tuple2):
    if len(tuple1) != len(tuple2):
        raise ValueError("tuples are of not same length")

    return tuple(a+b for a,b in zip(tuple1, tuple2))

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
output_tuple = tuple_elementwise_sum(tuple1, tuple2)
print(output_tuple)  # Expected output: (5, 7, 9)
```

Write a function that takes a tuple of strings and concatenates them, separating each string with a space.

Example

    input_tuple = ('Hello', 'World', 'from', 'Python')
    output_string = concatenate_strings(input_tuple)
    print(output_string)  # Expected output: 'Hello World from Python'




```python
def concatenate_strings(input_tuple):
    return ' '.join(input_tuple)

input_tuple = ('Hello', 'World', 'from', 'Python')
output_string = concatenate_strings(input_tuple)
print(output_string)  # Expected output: 'Hello World from Python'
```
