```python
import threading
import time

class ThreadCounter:

    def __init__(self):
        self.counter = 0
        # self.lock = threading.Lock()

    def count(self, thread_no):
        while True:
            self.counter += 1
            print(f"Thread {thread_no}: increased count to: {self.counter}")
            time.sleep(1)
            print(f"Thread {thread_no}: done some work, count is now: {self.counter}")


tc = ThreadCounter()
for i in range(20):
    t = threading.Thread(target=tc.count, args=(i,))
    t.start()
```
